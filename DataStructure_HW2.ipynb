## II.	做计时实验，比较list的pop()和pop(0)的性能 

```python
import time as t
import matplotlib.pyplot as plt
import numpy as np

def timer1(length):
    #a = np.arange(0,length,1)
    a = [i for i in range(length)]
    start = t.perf_counter()
    a.pop(0)
    end = t.perf_counter()
    return (end - start)

def timer2(length):
    #a = np.arange(0,length,1)
    a = [i for i in range(length)]
    start = t.perf_counter()
    a.pop()
    end = t.perf_counter()
    return (end - start)

def main():
    global result1,result2,length
    result1,result2 = [],[]
    length = np.arange(1,100000,100)
    for i in length:
        result1.append(timer1(i))
        result2.append(timer2(i))
    plt.figure(figsize = (8,5))
    plt.plot(length,result1,label='pop(0)')
    plt.plot(length,result2,color='red',linewidth=1.0,linestyle='--',label='pop()')
    plt.xlabel('length of list')
    plt.ylabel('time')
    plt.legend()
main()
```

## III.做计时实验，比较list和dict的del操作符性能，即 del lst[i]和del dic[key]

```python
import time as t
import matplotlib.pyplot as plt
import numpy as np

def timer1(length):
    a = [i for i in range(length)]
    start = t.perf_counter()
    del a[0]
    end = t.perf_counter()
    return (end - start)

def timer2(length):
    a = [i for i in range(length)]
    dic = {}
    for i in range(length):
        dic[i] = i
    start = t.perf_counter()
    del dic[0]
    end = t.perf_counter()
    return (end - start)

def main():
    #global result1,result2,length
    result1,result2 = [],[]
    length = np.arange(1,100000,100)
    for i in length:
        result1.append(timer1(i))
        result2.append(timer2(i))
    plt.figure(num = 3, figsize = (8,5))
    plt.plot(length,result1,label = 'del list[0]')
    plt.plot(length,result2,color='red',linewidth=1.0,linestyle='--',label = 'del dict[0]')
    plt.xlabel('length of list or dict')
    plt.ylabel('time')
    plt.legend()
    plt.show()
main()
```
